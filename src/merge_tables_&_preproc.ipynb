## Start merges

```python
# read results dataframe
results_df = pd.read_csv(config['raw']['results']['path'],sep=",")
results_df.columns = [str(col) + '_results' for col in results_df.columns]

#read results dataframe
status_df = pd.read_csv(config['raw']['status']['path'],sep=",")
status_df.columns = [str(col) + '_status' for col in status_df.columns]


results_df= results_df.merge(status_df,left_on="statusId_results",right_on="statusId_status")
print(results_df.shape)
print(results_df.columns)


qualy_df = pd.read_csv(config['raw']['qualifying']['path'],sep=",")
qualy_df.columns = [str(col) + '_qualy' for col in qualy_df.columns]
results_df= results_df.merge(qualy_df,left_on=["raceId_results","driverId_results","constructorId_results"],right_on=["raceId_qualy","driverId_qualy","constructorId_qualy"],how="left")
print(results_df.shape)
print(results_df.columns)
```

```python
race_df = pd.read_csv(config['raw']['races']['path'],sep=",")
race_df.columns = [str(col) + '_race' for col in race_df.columns]
results_df= results_df.merge(race_df[["raceId_race","year_race","round_race","date_race","circuitId_race"]],left_on=["raceId_results"],right_on=["raceId_race"],how="left")
print(results_df.shape)
print(results_df.columns)
```

```python
circuit_df = pd.read_csv(config['raw']['circuits']['path'],sep=",")
circuit_df.columns = [str(col) + '_circuit' for col in circuit_df.columns]
results_df= results_df.merge(circuit_df[["circuitId_circuit","location_circuit","country_circuit"]],left_on=["circuitId_race"],right_on=["circuitId_circuit"],how="left")
print(results_df.shape)
print(results_df.columns)
```

```python
constr_res_df = pd.read_csv(config['raw']['constructor_results']['path'],sep=",")
constr_res_df.columns = [str(col) + '_contr_res' for col in constr_res_df.columns]
results_df= results_df.merge(constr_res_df,left_on=["raceId_results","constructorId_results"],right_on=["raceId_contr_res","constructorId_contr_res"],how="left")
print(results_df.shape)
print(results_df.columns)
```

```python
constr_std_df = pd.read_csv(config['raw']['constructor_standings']['path'],sep=",")
constr_std_df.columns = [str(col) + '_constr_std' for col in constr_std_df.columns]
results_df= results_df.merge(constr_std_df,left_on=["raceId_results","constructorId_results"],right_on=["raceId_constr_std","constructorId_constr_std"],how="left")
print(results_df.shape)
print(results_df.columns)
```

```python
constr_df = pd.read_csv(config['raw']['constructors']['path'],sep=",")
constr_df.columns = [str(col) + '_constr' for col in constr_df.columns]
results_df= results_df.merge(constr_df[["constructorId_constr","constructorRef_constr","nationality_constr"]],left_on=["constructorId_results"],right_on=["constructorId_constr"],how="left")
print(results_df.shape)
print(results_df.columns)
```

```python
driver_df = pd.read_csv(config['raw']['drivers']['path'],sep=",")
driver_df.columns = [str(col) + '_driver' for col in driver_df.columns]
results_df= results_df.merge(driver_df,left_on=["driverId_results"],right_on=["driverId_driver"],how="left")
print(results_df.shape)
print(results_df.columns)
```

```python
driver_st_df = pd.read_csv(config['raw']['driver_standings']['path'],sep=",")
driver_st_df.columns = [str(col) + '_driver_st' for col in driver_st_df.columns]
results_df= results_df.merge(driver_st_df,left_on=["raceId_results","driverId_results"],right_on=["raceId_driver_st","driverId_driver_st"],how="left")
print(results_df.shape)
```

```python
lap_df = pd.read_csv(config['raw']['lap_times']['path'],sep=",")
lap_df.columns = [str(col) + '_lap' for col in lap_df.columns]
results_df= results_df.merge(lap_df,left_on=["raceId_results","driverId_results"],right_on=["raceId_lap","driverId_lap"],how="right")
results_df=results_df.drop(["time_lap","raceId_lap","driverId_lap"],axis=1)
print(results_df.shape)
```

```python
results_df=results_df[['date_race','year_race', 'round_race', 'location_circuit', 'country_circuit',
            'driverRef_driver','dob_driver', 'nationality_driver','points_driver_st', 'position_driver_st','wins_driver_st',
            'number_results', 'grid_results', 'position_results','points_results', 'laps_results', 'milliseconds_results', 'fastestLap_results', 'rank_results','fastestLapTime_results', 'fastestLapSpeed_results', 'status_status', 
            'constructorRef_constr','points_contr_res','points_constr_std', 'position_constr_std', 'wins_constr_std','nationality_constr',
            'number_qualy','position_qualy', 'q1_qualy', 'q2_qualy', 'q3_qualy', 'resultId_results', 'raceId_results', 'driverId_results','constructorId_results',
            ]]#'milliseconds_lap','lap_lap','position_lap']]
results_df.head()
```

```python
results_df['date_race'] = pd.to_datetime(results_df['date_race'], format='%Y/%m/%d')
results_df['dob_driver'] = pd.to_datetime(results_df['dob_driver'], format='%Y/%m/%d')
```

## Analyze value for columns
Now after merges we need to understand witch is the:
1. NaN precence/manage it
2. Tipical value for each column
3. convert date in date format
4. convert time in millisecond

```python
df=results_df[results_df['year_race']>=2004]
```

```python
def time_to_ms(time_str):
    try:
        minutes, seconds = time_str.split(':')
        seconds, milliseconds = seconds.split('.')
        time = datetime.time(minute=int(minutes), second=int(seconds), microsecond=int(milliseconds)*1000)
        return (time.hour * 3600 + time.minute * 60 + time.second) * 1000 + time.microsecond // 1000
    except:
        return -1
for col in ['q1_qualy','q2_qualy','q3_qualy','fastestLapTime_results']:
    df[col]=df[col].astype(str)
    df[col] = df[col].apply(time_to_ms).astype(int)
```

```python
dtype={
    'date_race':datetime.datetime,
    'year_race':int,
    'round_race':int,
    'location_circuit':str,
    'country_circuit':str,
    'driverRef_driver':str,
    'dob_driver':datetime.datetime,
    'nationality_driver':str,
    'points_driver_st':float,
    'position_driver_st':int,
    'wins_driver_st':int,
    'number_results':int,
    'grid_results':int, 
    'position_results':int,
    'points_results':int,#<-TOFIX 
    'laps_results':int,
    'milliseconds_results':float,
    'fastestLap_results':float,
    'rank_results':int,
    'fastestLapTime_results':int,
    'fastestLapSpeed_results':float, 
    'status_status':str, 
    'constructorRef_constr':str,
    'points_contr_res':int, 
    'points_constr_std':float,
    'position_constr_std':int,
    'wins_constr_std':int, 
    'nationality_constr':str,
    'number_qualy':int,
    'position_qualy':int,
    'q1_qualy':int,
    'q2_qualy':int,
    'q3_qualy':int,
    'resultId_results':int, 
    'raceId_results':int,
    'driverId_results':int,
    'constructorId_results':int,
    'lap_lap':int,
    'position_lap':int,
    'milliseconds_lap':int  
}

def converter_int(x):
    try:
        return int(x)
    except:
        return -1
def converter_float(x):
    try:
        return float(x)
    except:
        return -1
def converter_str(x):
    try:
        return str(x)
    except:
        return "N"
    
for col in ['position_results','position_lap','lap_lap','milliseconds_lap','milliseconds_results','fastestLap_results','position_qualy','rank_results','position_constr_std','wins_constr_std','number_qualy']:
    print(col)
    df[col] = df[col].apply(converter_int)
    print(df[col].value_counts())
for col in ['fastestLapSpeed_results']:
    print(col)
    df[col] = df[col].apply(converter_float)
    print(df[col].value_counts())

df['points_driver_st']=df['points_driver_st'].fillna(0)
df['position_driver_st']=df['position_driver_st'].fillna(0)
df['wins_driver_st']=df['wins_driver_st'].fillna(0)
df['points_contr_res']=df['points_contr_res'].fillna(0)
df['points_constr_std']=df['points_constr_std'].fillna(0)
```
